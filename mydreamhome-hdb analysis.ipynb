```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# To determine the size of the chart and the quality.
fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

# Read data from a file
data_1 = np.genfromtxt('resale-transactions-by-flat-type-based-on-registered-cases.csv',
                    delimiter = ',',
                    dtype=None,
                    encoding=None)

#print(data_1.shape)
#print(data_1)
print("***Resale Transactions By Flat Type Based On Registered Cases***\n")
print("There are {} rows and {} columns in this dataset\n".format(data_1.shape[0], data_1.shape[1]))

# To extract the title name from the csv
bt = data_1[0, :] # all the title names
bt0 = data_1[0,0] # financial_year
bt1 = data_1[0,1] # flat_type
bt2 = data_1[0,2] # resale_transactions
#print(bt) 


# To change the class type from array to string
fin_yr = type(np.array2string(bt0, precision=2, separator=',', suppress_small=True)) # financial_year
ft = type(np.array2string(bt1, precision=2, separator=',', suppress_small=True)) # flat_type
rtran = type(np.array2string(bt2, precision=2, separator=',', suppress_small=True)) # resale_transactions

# To check the title name if its numeric
fin_yr1 = bt0.isnumeric()
ft1 = bt1.isnumeric()
rtran1 = bt2.isnumeric()

print(f'The names of the columns are: \n - {bt0} {fin_yr} isnumeric: {fin_yr1} \n - {bt1} {ft} isnumeric: {ft1} \n - {bt2} {rtran} isnumeric: {rtran1} \n')

# To calculate the unique values in the following columns
datau1 = len(np.unique(data_1[1:,0])) 
datau2= len(np.unique(data_1[1:,1]))
datau3 = len(np.unique(data_1[1:,2]))

print(f'{datau1} unique values in {bt0} column \n{datau2} unique values in {bt1} column \n{datau3} unique values in {bt2} column \n')

# To extract the data 
room = data_1[[88,89,90,91,92,93], 1:2 # room type
x1 = room.astype(str)
#print(x1)

noOfTransactions = data_1[[88,89,90,91,92,93], 2:] # no of transaction for the room type.
y1 = noOfTransactions.astype(int)
#print(y1)

# To change the int values to percentage
totalroom = np.sum(y1)
#print(totalroom)
threeroom = ((6023/totalroom) * 100)
fourroom = ((10401/totalroom) * 100)
fiveroom = ((6005/totalroom) * 100)
Execroom = ((1818/totalroom) * 100)

# To label the values that was chosen
labels = ['3 room', '4 room', '5 room', 'Executive']
sizes = [threeroom, fourroom, fiveroom, Execroom]
split = [0, 0.2, 0, 0]
mycolors = ['#43CD80', '#2E8B57', '#4EEE94', '#54FF9F']

# To create the pie chart
ax.pie(sizes,
       explode = split,
       labels = labels,
       colors = mycolors,
       autopct = '%1.2f%%',
       shadow = True,
       startangle = 50)

ax.axis('equal')
ax.set_title(label='3-room/4-room/5-room/Executive Flats sold in 2019', fontsize=14, pad=20)
plt.legend(loc="lower right", bbox_to_anchor=(0, -0.15), ncol= 1, title = 'Types of flats:', )
plt.savefig('resale_transactions.png')
plt.show()
```
