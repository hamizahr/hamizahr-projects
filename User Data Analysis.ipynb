### Q1: How many users are Male vs Females?

```python
# Create container to update values
male = 0
female = 0

# Iterate through index of df
for idx in df.index:
    
    # Extract and name variables based on index & columns
    gender = df['gender'][idx]
    
    #Test the logic if gender = male
    if gender == "Male": #True
        male += 1               # Update the values
            
    else: # False
        female += 1            # Update the values

print("Male:", male, "vs", "Female:", female)
```

### Q2: Which country has the most users?

```python
# Create dict to update values
results = {}

# Iterate through index of df
for idx in df.index:
    
    # Extract and name variables based on the index & columns
    citizenship = df['citizenship'][idx]
    #print(citizenship)
    
    # To check if the values are in the results container if not it will be updated
    if citizenship not in results:
        results[citizenship] = 1
    else:
        results[citizenship] += 1

print(results)
```

```python
most_users_country = ''
most_users = 0

for country, count in results.items():
    if count > most_users:
        
        most_users_country = country
        most_users = count
        
print("The country with the most users is", most_users_country, "with", most_users, "users.")
```

### Q3: How many users have invalid identification value? The correct identification value must have 10 characters.

```python
#Create container to update values
ID = 0

# iterate through index of df
for idx in df.index:
    
    # Extract and name variables based on idx & columns
    identification = str(df['identification'][idx])

    if len(identification) < 10:
        #print(identification, len(identification))
        ID += 1
        
print("The number of users that have invalid identification value:", ID)
```

### Q4: How many users have valid and invalid mobile phone numbers? Use the country of birth to check. If user country of birth is Singapore and mobile number is +65 then valid else invalid.

```python
#Create container to update values
valid = 0
invalid = 0

# iterate through index of df
for idx in df.index:
    
    # Extract and name variables based on idx & columns
    mobile_phone = str(df['mobile_phone'][idx])
    mobile_home = str(df['mobile_home'][idx])
    country_birth = df['country_birth'][idx]
    
    #Test the logic if country_birth = Singapore and starts with "65"
    if country_birth == "Singapore": #True
        if mobile_phone.startswith("65") & mobile_home.startswith("65"):
            valid += 1               # Update the values
            
    #Test the logic if country_birth != Singapore and starts with "65"
    else: # False
        if mobile_phone.startswith("65") & mobile_home.startswith("65"):
            invalid += 1            # Update the values
            
print("The number of users born in Singapore:", valid)
print("The number of users not born in Singapore:", invalid)
```

### Q5: Replace driver license 1, 2, 3 to the following: 1 - class A, 2 - class B, 3 - class C [Do not use search, find and replace]

```python
# Iterate through index of df
for idx in df.index:
    
    # Extraction of the driver license value and assign as a variable
    driver_license = df['driver_license'][idx]
    
    # Check for condition
    if driver_license == 1: #True
        
        # Replace the original df value based on df[col][idx]
        df['driver_license'][idx] = "class A"
        
    elif driver_license == 2:
        df['driver_license'][idx] = "class B"
        
    else:
        df['driver_license'][idx] = "class C"
        
print(df)
```
